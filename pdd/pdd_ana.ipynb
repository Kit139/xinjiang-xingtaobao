```python
import pandas as pd

df = pd.read_csv('pdd_xinjiangxing.csv')
# df
```

```python
df.dropna(subset=['sales'],inplace=True)
df['prices'] = df['prices'].str.strip('券后¥')
df['sales'] = df['sales'].str.strip('已拼').str.strip('总售').str.strip('件').str.strip('+')
# df['sales'] = df['sales'].str.replace('万','0000')
df['sales'] = df.apply(lambda x: str(int(float(x['sales'].replace('万',''))*10000)) if "万" in x['sales'] else x['sales'], axis=1)
df[[]]
```
